```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows 한글 폰트
matplotlib.rcParams['font.size'] = '10' # font size
matplotlib.rcParams['axes.unicode_minus'] = False # - 기호 깨짐 현상을 해결함
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df.head(3)
```

```python
x = df['이름'].head(5)
y = df['국어'].head(5)
y2 = df['영어'].head(5)
y3 = df['수학'].head(5)
```

```python
plt.plot(x,y,label='국어',marker='D')
plt.plot(x,y2,label='영어',marker='o',ls='--')
plt.plot(x,y3,label='수학',marker='s',ls='-.')
plt.legend(loc='upper center',ncol=3)

for i,txt in enumerate(y3):
    plt.text(x[i],y3[i]+1,txt,ha='center',color='black')
```

```python
# 패턴넣기
bar =plt.bar(x,y,label='국어',width=0.2, alpha=0.5)
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('..')
bar[3].set_hatch('-')
bar[4].set_hatch('|')
plt.legend(loc='upper right')
# plt.xticks(rotation=90)


for i,txt in enumerate(y):
    plt.text(x[i],y[i]+0.5,txt,ha='center',color='black')
```
