```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows 한글 폰트
matplotlib.rcParams['font.size'] = '10' # font size
matplotlib.rcParams['axes.unicode_minus'] = False # - 기호 깨짐 현상을 해결함
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
# 이름, 국어 성적을 가지고 선 그래프를 출력
x = df['이름']
y = df['국어']

plt.bar(x,y,label='국어점수')
plt.legend(loc='upper center')
plt.title('빅데이터반 성적그래프')
```

```python
# chipotle.tsv 에서 주문 중 가장 많이 주문한 메뉴 10개 
import pandas as pd
df = pd.read_csv('../z20_data/chipotle.tsv',sep='\t')
df
```

```python
filt = df.groupby('item_name')[['quantity']].sum().sort_values('quantity',ascending=False).head()

import numpy as np

x = filt.index
y = filt.values

plt.plot(x,y,label='총 주문 횟수')
plt.plot(x,y,'bo')
plt.xticks(np.arange(0,10,1),size=8)

# 주문 횟수가 몇번인지
item =df['item_name'].value_counts().head()
x2 = item.index 
y2 = item.values
plt.plot(x2,y2,label='주문한 메뉴 개수')
plt.plot(x2,y2,'ro')
plt.xticks(rotation=90)
plt.legend(loc=1)
```
