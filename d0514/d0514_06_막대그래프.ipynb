```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows 한글 폰트
matplotlib.rcParams['font.size'] = '10' # font size
matplotlib.rcParams['axes.unicode_minus'] = False # - 기호 깨짐 현상을 해결함
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
# 합계 컬럼 생성해서 막대그래프 가로,세로
df['합계'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df
```

```python
x = df['이름']
y = df['합계']

plt.bar(x,y,label='합계')
plt.legend(loc='upper center')
plt.ylim(0,500)
for i,txt in enumerate(y):
    plt.text(x[i],y[i]+3,txt,ha='center',color='r')
```

```python
x = df['이름']
y = df['합계']

plt.barh(x,y,label='합계',color='y')
plt.legend(loc='right')
plt.xlim(0,500)
for i,txt in enumerate(y):
    plt.text(y[i]+2,x[i],txt,color='g')
```
