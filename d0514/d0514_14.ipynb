```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows 한글 폰트
matplotlib.rcParams['font.size'] = '10' # font size
matplotlib.rcParams['axes.unicode_minus'] = False # - 기호 깨짐 현상을 해결함
import numpy as np
```

```python
# 막대그래프 score.xlsx 합계 점수를 뽑아서 그래프로
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df.head(3)
```

```python
df['합계'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df
```

```python
x = df['이름']
y = df['합계']

plt.bar(x,y,label='합계',width=0.5,color=['r','orange','#fce803','g','b','#222eba','m','black'])
plt.legend(loc=9)
for i,txt in enumerate(y):
    plt.text(x[i],y[i]+3,txt,ha='center')
plt.ylim(0,500)
plt.title('합계')
plt.grid(axis='y',alpha=0.5,linestyle='--')
plt.show()
```
