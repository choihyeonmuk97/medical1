```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows 한글 폰트
matplotlib.rcParams['font.size'] = '10' # font size
matplotlib.rcParams['axes.unicode_minus'] = False # - 기호 깨짐 현상을 해결함
```

```python
x=[1,2,3]
y=[2,4,8]


plt.plot(x,y,label='국어',linestyle=':',marker='o')
plt.legend(loc=2)
plt.title('그래프')
```

```python
import numpy as np
x=[1,2,3]
y=[2,4,8]


# plt.figure(dpi=200) # 해상도 크기 조절
plt.plot(x,y,label='국어',linestyle=':',marker='o')
plt.legend(loc=2)
for i,txt in enumerate(y):
    plt.text(x[i],y[i]+0.3,txt,ha='center',color='b')
plt.yticks(np.arange(0,11,1))
# plt.savefig('g2.png',dpi=200) # dpi : 해상도 설정

plt.show()
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df.head(3)
```

```python
# 키 그래프 출력 (이름,키) / y축 170~210(10단위)
df['키']
x = df['이름']
y = df['키']

plt.plot(x,y,label='키')
plt.plot(x,y,'ro')
plt.legend(loc=1)
plt.yticks([150,160,170,180,190,200,210])
for i,txt in enumerate(y):
    plt.text(x[i],y[i]-3.5,txt, ha = 'center',color='g')
plt.title('키 그래프')
```

```python
# 키 그래프 출력 (이름,키) / y축 170~210(10단위)
df['키']
x = df['이름']
y = df['키']

plt.barh(x,y,label='키') # barh -> 막대그래프가 가로로 나타난다.
plt.legend(loc='upper right')
# plt.yticks([150,160,170,180,190,200,210])
# plt.ylim(165,210)
plt.xlim(165,210)
for i,txt in enumerate(y):
    plt.text(y[i]+2,x[i],txt, ha = 'center',color='g')
plt.title('키 그래프')
```
