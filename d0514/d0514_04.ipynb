```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows 한글 폰트
matplotlib.rcParams['font.size'] = '10' # font size
matplotlib.rcParams['axes.unicode_minus'] = False # - 기호 깨짐 현상을 해결함
```

```python
x=[1,2,3]
y=[5,4,8]


# plt.plot(x,y,label='국어',linewidth=3, marker='o', mec='white', mfc='red') # 약어 존재
plt.plot(x,y,label='국어',linewidth=3, marker='o', markeredgecolor='white', markerfacecolor='red', alpha=0.7, ls=':')
plt.title('막대그래프')
plt.legend(loc='lower right') # 위치점으로도 지정 가능 ex) loc=(0.5,0.5)
plt.xticks([1,2,3])
plt.yticks([5,6,7,8,9,10,11,12])
```
