```python
# dropna - NaN 지우기
# axis : index, columns
# how : any, all
df.dropna(axis='columns',inplace=True) # NaN 데이터가 있는 컬럼을 전부 삭제함
```

```python
df.dropna(axis='columns', how='all', inplace=True) # NaN이 있는 컬럼에서 모든 컬럼이 NaN 인 경우 컬럼 전체 삭제
```

```python
df.dropna(axis='columns',how='any',inplace=True) # NaN이 들어있는 컬럼에서 1개라도 NaN이 포함되어 있으면 컬럼을 삭제함
```
