```python
import pandas as pd
df = pd.read_csv('../z20_data/chipotle.tsv',sep='\t')
df
```

```python
len(df['item_name'].unique()) # 중복 제거 후 메뉴의 개수 세기
```

```python
# 각 아이템 별 가격을 구하기
# 아이템 별 주문개수가 1개인 것의 가격?
filt = df[df['quantity'] == 1]

# 아이템 별 최저가격
a = filt.groupby('item_name').min().sort_values('item_price')
a['item_price']
```

```python
# 형변환 다시 ㅋ
df['item_price'] =df['item_price'].str.replace('$','').astype(float)
```

```python
# order_id 별 item_price의 합이 가장 높은것을 역순정렬 하 고
# quantity (주문횟수), item_price를 출력하라는 것이었음

a = df.groupby('order_id').sum().sort_values('item_price',ascending=False)
a[['quantity','item_price']]
```

```python
# 주문 당 개수가 가장 많은 상위 5개 출력
df.groupby('order_id').sum().sort_values('quantity',ascending=False)[['quantity','item_price']].head()
```

```python
# 메뉴별 주문 수량
# df['item_name'].value_counts()[:10] -> 그룹바이 안하고 수량뽑기
df.groupby('item_name').size().sort_values(ascending=False)[:10]
```

```python
# 출력 이쁘게..
list(df['item_name'].value_counts()[:10].items()) 

item_count = df['item_name'].value_counts()[:10]

for i,(val,cnt) in enumerate(item_count.items()):
    print(f'Top {i+1} :',val,'-',cnt)
```

```python
df['item_price'] = df['item_price'].str.replace('$','').astype(float)
```

```python
# 주문 당 평균 계산 금액
# df.groupby('order_id').size()
df.groupby('order_id')['item_price'].mean()
```

```python
# 한 주문에 20$ 이상 주문한 id 출력
order_id_group = df.groupby('order_id').sum()
df.groupby('order_id')['item_price'].sum() > 20

order_id_group.loc[df.groupby('order_id')['item_price'].sum() > 20,['item_price']]
```

```python
# Veggie Salad Bowl 메뉴가 몇번 주문 되었는지
df[df['item_name'] == 'Veggie Salad Bowl'].count()['quantity']
```

```python
# 저 메뉴 가격 역순정렬
df[df['item_name'] == 'Veggie Salad Bowl'].sort_values('item_price',ascending=False)
```

```python
# Chicken Bowl 2개 이상 주문한 주문 횟수?
filt = (df['item_name'] == 'Chicken Bowl') & (df['quantity'] >= 2)
df.loc[filt,'quantity'].count()
```

```python
# 치킨보울 총 주문 개수 (quantity, 수량)
filt = (df['item_name'] == 'Chicken Bowl') & (df['quantity'] >= 2)
df.loc[filt,'quantity'].sum()
```
