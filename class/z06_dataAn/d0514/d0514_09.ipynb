```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows 한글 폰트
matplotlib.rcParams['font.size'] = '10' # font size
matplotlib.rcParams['axes.unicode_minus'] = False # - 기호 깨짐 현상을 해결함
import numpy as np
```

```python
import pandas as pd
df = pd.read_csv('../z20_data/stu_score1.csv')
df.head(3)
```

```python
h_df = df.head(20)
x = df['id'].head(20)
y = df['total'].head(20)
y1 = df['kor'].head(20)
y2 = df['eng'].head(20)
y3 = df['math'].head(20)
```

```python
plt.bar(x,y)
plt.ylim(150,300)
plt.xticks(rotation=60)
plt.show()
```

```python
h_df.shape[0]
```

```python
idx =np.arange(h_df.shape[0])
```

```python
plt.figure(figsize=(15,5))
plt.bar(idx-0.3,y1,width=0.3)
plt.bar(idx,y2,width=0.3)
plt.bar(idx+0.3,y3,width=0.3)
for i,txt in enumerate(y2):
    plt.text(idx[i],y2[i]+0.5,txt,ha='center')


plt.xticks(idx,x,rotation=60)

plt.show()
```
