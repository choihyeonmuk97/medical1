```python
import pandas as pd
df =  pd.read_csv('../z20_data/drinks.csv')
df.head(20)
```

```python
df.fillna('NN',inplace=True)
```

```python
df.head(10)
```

```python
# 대륙별 최소, 최대, 평균, 합계
df.groupby('continent')[['beer_servings','spirit_servings','wine_servings','total_litres_of_pure_alcohol']].agg(['min','max','mean','sum'])
```

```python
# 대륙 중 아프리카 국가별 퍼센트로 출력
# df['continent'].count()
len(df.groupby('continent').value_counts(normalize=True).loc['AF'] )
df.groupby('continent').value_counts(normalize=True).loc['AF'] 
```

```python
df.groupby('continent')['beer_servings'].sum()
```

```python
# 1. 전체 평균 소비량
beer_mean = df['beer_servings'].mean()
beer_mean
```

```python
# 전체 평균보다 맥주 소비량이 높은 대륙
# 1. 전체 평균 소비량
beer_mean = df['beer_servings'].mean()
# 2. 대륙별 평균 소비량
continent_over_mean = df.groupby('continent')['beer_servings'].mean()
# continent_over_mean

continent_over_mean[continent_over_mean >= beer_mean]
```

```python
# total_litres_of_pure_alcohol 가 전체 평균보다 높은 대륙
total_mean = df['total_litres_of_pure_alcohol'].mean()
total_mean
```

```python
alcohol_mean = df.groupby('continent')['total_litres_of_pure_alcohol'].mean() 
alcohol_mean[alcohol_mean>=total_mean]
```
