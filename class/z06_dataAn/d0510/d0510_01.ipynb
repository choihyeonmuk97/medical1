```python
# 출생파일 skiprows=2 use B:M
import pandas as pd
df = pd.read_excel('../z20_data/20122022_출생.xlsx',skiprows=2, nrows=3 ,usecols='B:M',index_col=0)
df.index.name = '목록'
df.iloc[0] = df.iloc[0].str.replace(',','').astype(int)

df['합계'] = 0 # 컬럼 추가

df.iloc[0,-1] = df.iloc[0].sum() # 추가된 합계 컬럼에 데이터 추가
df
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
import numpy as np
df.iloc[2,4] = np.nan
df.iloc[3,5] = np.nan
df.iloc[4,[3,4]] = np.nan
df
```

```python
df.fillna(0,inplace=True)
df
```

```python
df['국어'] = df['국어'].astype(int)
```

```python
# 총합 컬럼 생성 후 국영수사과 합계 점수를 총합 컬럼에 넣기
df['총합'] = 0
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df
# df
```

```python
df['결과'] = '불합격'
# 총합이 300점 이상이면 합격 이하 불합격
filt = df['총합'] >= 300
df.loc[filt,'결과'] = '합격'

# 400점 이상은 우수
df.loc[filt,'결과'] = '우수'
```

```python
# 컬럼 위치 변경
cols = list(df.columns)
df = df[cols[0:8]+cols[9:]+[cols[8]]]
```
