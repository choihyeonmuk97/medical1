```python
import numpy as np
import pandas as pd
from tensorflow import keras
```

```python
(train_input,train_target),(test_input,test_target) = keras.datasets.mnist.load_data()
```

```python
train_data = pd.read_csv('../mnist/train.csv',header=None) # train
test_data = pd.read_csv('../mnist/t10k.csv',header=None) # test
test_data.head()
```

```python
# train_input = train_data.drop(0,axis=1)
train_input = train_data.iloc[:,1:]
# test_input = test_data.drop(0,axis=1)
test_input = test_data.iloc[:,1:]
```

```python
train_input
```

```python
train_target = train_data[0]
test_target = test_data[0]
```

```python
train_target
```

```python
# 0~1 까지의 값으로 변경
train_input = train_input/255
test_input = test_input/255
```

```python
from sklearn.model_selection import train_test_split
train_input,val_input,train_target,val_target = train_test_split(
    train_input,train_target,test_size=0.2,random_state=42
)
```

```python
train_target.value_counts().sort_index()
```

```python
# dense층 만들기
dense = keras.layers.Dense(10,activation='softmax',input_shape=(784,))
model = keras.Sequential([dense])
model.compile(loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.fit(train_input,train_target,epochs=5)
model.evaluate(val_input,val_target)
```
